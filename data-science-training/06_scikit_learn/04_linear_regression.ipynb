```
import sklearn
```

```
import numpy as np
```

```
X = np.arange(0,50,0.5)
noise = 10 * np.random.randn(100)
y = 2* X +100 + noise
X = X.reshape(-1,1)

print('Shape X:',X.shape)
print('Shape y:',y.shape)
```

```
from sklearn.model_selection import train_test_split

X_train, X_test , y_train, y_test = train_test_split(X,y,test_size=0.25)
```

```
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()

plt.figure(figsize=(7,5))
plt.scatter(X_train,y_train,c='b',label="Training set")
plt.scatter(X_test,y_test,c='g',label="Test set")
plt.xlabel('X')
plt.ylabel('y')
_ = plt.legend()
```

```
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X_train,y_train)

print(lin_reg.coef_)
print(lin_reg.intercept_)
```

```
plt.figure(figsize=(7,5))
plt.title('Linear Regression - one variable')
plt.scatter(X_train,y_train,c='b',label='Training set')
plt.plot(X, lin_reg.intercept_ + lin_reg.coef_ * X, c='red')
plt.xlabel('X')
plt.ylabel('y')
_ = plt.legend()
```

```
y_pred = lin_reg.predict(X_test)
plt.figure(figsize=(7,5))
plt.title('Linear Regression - one variable')
plt.scatter(X_test,y_test,c='g',label='Test set')
plt.plot(X, lin_reg.intercept_ + lin_reg.coef_ * X, c='red')
plt.xlabel('X')
plt.ylabel('y')
_ = plt.legend()
```

```
lin_reg.score(X_test,y_test)
```
