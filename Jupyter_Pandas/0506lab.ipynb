```python
import pandas as pd
import numpy as np
import time
from datetime import datetime 
from datetime import timedelta
```

```python
df = pd.read_csv("C:/Users/user/Desktop/Python_Main/kurs_pandas/marathon_results_2016.csv",
                usecols=['Bib','40K','Half','Pace','Age','M/F','Country','State','City'],
                index_col = "Bib")
df.head(1000)
```

```python
df['40K'] = df['40K'].apply(pd.to_timedelta) 
df['Half'] = df['Half'].apply(pd.to_timedelta) 
df['TotalSeconds'] = df['40K'].apply(lambda x: timedelta.total_seconds(x)) 
df['HalfSeconds'] = df['Half'].apply(lambda x: timedelta.total_seconds(x))
```

```python
x = df.pivot_table(index = "Age", columns="M/F",values=["TotalSeconds","HalfSeconds"]).swaplevel(axis=1)
```
