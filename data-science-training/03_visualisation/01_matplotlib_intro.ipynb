```
import matplotlib
```

```
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```
fig = plt.figure()
fig.show()
```

```
_ = plt.plot([0,1],[0,3])
```

```
plt.plot([0,1],[0,3],label='line',color='green')
plt.xlabel('axis x')
plt.ylabel('axis y')
plt.title('Chart')
_ = plt.legend()
```

```
plt.plot([0,1,2],[0,3,2],label='line',color='green')
plt.xlabel('axis x')
plt.ylabel('axis y')
plt.title('Chart')
_ = plt.legend()
```

```
x = np.arange(-3,3,0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)
plt.plot(x,x_sin)
_ = plt.plot(x,x_cos)
```

```
plt.subplot(121)
plt.plot([0,1,2],[3,2,5])
plt.subplot(122)
plt.plot([0,1,2],[3,2,4],'r--')
```

```
fig =plt.figure(figsize=(12,8))

plt.subplot(131)
plt.plot([0,1,2],[3,2,5])
plt.subplot(132)
plt.plot([0,1,2],[3,2,4],'r--')
plt.subplot(133)
plt.plot([0,1,2],[3,1,4],'ro')
```

```
plt.bar(x=[0,1,2,3],height=[1.2,3.2,1.2,5.2])
```

```
plt.bar(x=['a','b','c','d'],height=[1.2,3.2,1.2,5.2])
```

```
plt.barh(y=['a','b','c','d'],width=[1.2,3.2,1.2,5.2])
```
